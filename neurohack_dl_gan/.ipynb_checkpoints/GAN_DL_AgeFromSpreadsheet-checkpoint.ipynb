```python
import os, sys, time, h5py, zipfile
import numpy as np
import pandas as pd
```

```python
behav_data = pd.read_csv('/home/anna/Scrivania/NeuroHackademy2019/unrestricted_annatruzzi91_8_5_2019_13_21_16.csv')
```

```python
behav_data.keys()
```

```python
age = behav_data[['Subject','Age']]
```

```python
min_age_vector=[]
max_age_vector=[]
for item in age['Age']:
    if '+' in item:
        min_age_vector.append(int(item[0:2]))
        max_age_vector.append(90)
    else:
        min_age_vector.append(int(item[0:2]))
        max_age_vector.append(int(item[-2:]))

min_age_vector=np.asarray(min_age_vector)
max_age_vector=np.asarray(max_age_vector)
```

```python
min_age_vector
```

```python
unique, counts = np.unique(min_age_vector, return_counts=True)
dict(zip(unique, counts))
agelist = unique
```

```python
age_category = np.zeros(shape=len(age['Age']))
```

```python
age['min_age'] = pd.Series(min_age_vector)
age['age_category'] = pd.Series(age_category)
```

```python
age
```

```python
age.loc[age['min_age'] == agelist[0],['age_category']] = 0
age.loc[age['min_age'] == agelist[1],['age_category']] = 1
age.loc[age['min_age'] == agelist[2],['age_category']] = 2
age.loc[age['min_age'] == agelist[3],['age_category']] = 3
```
